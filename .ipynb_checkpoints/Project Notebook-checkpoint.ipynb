### Importing Data

```python
# import data
por_raw = pd.read_csv('student-por.csv')
mat_raw = pd.read_csv('student-mat.csv')

# remove observations that contain students aged 18 and above
por_raw = por_raw[por_raw['age'] < 18]
mat_raw = mat_raw[mat_raw['age'] < 18]
```

### Predictive Analysis
#### Linear Regression - Predicting Final Grades (Math Class)

```python
x = pd.get_dummies(mat_raw[['school', 'sex', 'age', 'address', 'famsize', 'Pstatus', 'Medu', 'Fedu', 'Mjob', 'Fjob', 'reason', 'guardian', 'traveltime', 'studytime', 'failures', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet', 'romantic', 'famrel', 'freetime', 'goout', 'Dalc', 'Walc', 'health', 'absences']], 
                                 columns = ['school', 'sex', 'address', 'famsize', 'Pstatus', 'Medu', 'Fedu', 'Mjob', 'Fjob', 'reason', 'guardian', 'traveltime', 'studytime', 'failures', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet', 'romantic', 'famrel', 'freetime', 'goout', 'Dalc', 'Walc', 'health'], drop_first=True)
features = sm.add_constant(x)
target = mat_raw[['G3']]

model = sm.OLS(target, features)
result = model.fit()
result.summary()
```

```python
# keeping coefficients that were statistically significant in previous OLS summary
x = pd.get_dummies(mat_raw[['Medu', 'Mjob', 'failures', 'famsup', 'romantic', 'freetime', 'Walc']], 
                                 columns = ['Medu', 'Mjob', 'failures', 'famsup', 'romantic', 'freetime', 'Walc'], drop_first=True)

features = sm.add_constant(x)
target = mat_raw[['G3']]

model = sm.OLS(target, features)
result = model.fit()
result.summary()
```

```python
# creating sci-kit learn model and making predictions
X = sm.add_constant(pd.get_dummies(mat_raw[['Medu', 'Mjob', 'failures', 'famsup', 'romantic', 'freetime', 'Walc']], 
                                                 columns = ['Medu', 'Mjob', 'failures', 'famsup', 'romantic', 'freetime', 'Walc'], drop_first=True))
y= mat_raw[['G3']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=5)

model = LinearRegression().fit(X_train, y_train)
preds = model.predict(X_test)

# MSE
print('RMSE: ' + str(mean_squared_error(y_test, preds, squared=False)))

# R2
_, _, r_value, _, std_err = scipy.stats.linregress(y_test.to_numpy().squeeze(), preds.squeeze())
print('R2: ' + str(r_value**2))

preds = pd.Series(preds.ravel())
y_test = pd.Series(y_test.to_numpy().ravel())
dd = pd.DataFrame({'test_set':y_test, 'prediction':preds})
dd.plot.scatter(x='test_set', y='prediction',xticks=range(0,21,2), yticks=range(0,21,2));
```
